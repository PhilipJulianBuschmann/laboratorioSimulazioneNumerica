<span style="color:red">Include the calculation of the pressure, $P$, inside your Molecular Dynamics code: block averages, progressive average values and progresssive uncertainties.</span>

Remember that, via the **virial theorem**, the pressure in a Lennard-Jones system can be estimated by means of:

$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=40000              # Total number of throws
N=20                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N)      # [0,1,2,...,N-1]

numblockKin, valmedblockKin, progmeanKin, meanerrKin = np.loadtxt("MDNVE_MCNVT/output_ekin.datLiq", unpack = True);
numblockPot, valmedblockPot, progmeanPot, meanerrPot = np.loadtxt("MDNVE_MCNVT/output_epot.datLiq", unpack = True);
numblockTot, valmedblockTot, progmeanTot, meanerrTot = np.loadtxt("MDNVE_MCNVT/output_etot.datLiq", unpack = True);
numblockPres, valmedblockPres, progmeanPres, meanerrPres = np.loadtxt("MDNVE_MCNVT/output_pres.datLiq", unpack = True);
numblockTemp, valmedblockTemp, progmeanTemp, meanerrTemp = np.loadtxt("MDNVE_MCNVT/output_temp.datLiq", unpack = True);

#x*=L
plt.errorbar(x,progmeanKin,yerr=meanerrKin, label = 'Ekin')
plt.errorbar(x,progmeanPot,yerr=meanerrPot, label = 'Epot')
plt.errorbar(x,progmeanTot,yerr=meanerrTot, label = 'Etot')
plt.errorbar(x,progmeanPres,yerr=meanerrPres, label = 'pres')
plt.errorbar(x,progmeanTemp,yerr=meanerrTemp, label = 'temp')
plt.legend();
plt.xlabel('steps')
plt.ylabel('')
plt.grid(True)
plt.title('Liquid phase\n$Temp = 1.1$\n$Rho = 0.8$\n$N_{part} = 108$')
plt.show()
```
